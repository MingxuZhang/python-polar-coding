```python
from python_polar_coding.polar_codes.g_fast_ssc import GeneralizedFastSSCPolarCodec
from python_polar_coding.polar_codes.e_g_fast_ssc import EGFastSSCPolarCodec
from math import ceil
import pandas as pd
from collections import Counter
```

```python
# N in range 1024, ... 32768
N_range = [int(2 ** i) for i in range(10, 16)]
R_range = [1/8, 1/4, 1/2, 3/4, 7/8]
DSNR = 2.5
AF = 3
```

```python
data = list()
for N in N_range:
  for R in R_range:
    K = ceil(N * R)
    code = GeneralizedFastSSCPolarCodec(N=N, K=K, design_snr=DSNR, AF=AF)
    code_info = dict(
        type='G-Fast-SSC',
        N=code.N,
        K=code.K,
        subcodes=len(code.tree.leaves),
        **dict(Counter(l.type for l in code.tree.leaves))
    )
    data.append(code_info)

    exp_code = EGFastSSCPolarCodec(N=N, K=K, design_snr=DSNR, AF=AF)
    exp_code_info = dict(
        type='Exp-G-Fast-SSC',
        N=exp_code.N,
        K=exp_code.K,
        subcodes=len(code.tree.leaves),
        **dict(Counter(l.type for l in exp_code.tree.leaves))
    )
    data.append(exp_code_info)
```

```python
pd.DataFrame.from_dict(data)
```
